```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of plots

#### Line plots

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True);
```

#### Bar Chart

```python
months = ['Jan','Feb','Mar','Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months, revenue);
plt.title('Monthly Revenue Bar Chart');
plt.xlabel('Month');
plt.ylabel('Revenue');
```

#### Histogram

```python
marks50 = np.random.randint(0, 100, 50)
marks50
```

```python
plt.hist(marks50);
plt.xlabel('Marks');
plt.ylabel('Number of Students');
```

```python
plt.hist(marks50, bins=5);
```

#### Scatter plot

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 180, 145]
```

```python
plt.scatter(heights, weights);
plt.title('Heights vs Weights Scatter Plot');
plt.xlabel('Height (inches)');
plt.ylabel('Weights (pounds)');
```

## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();
```

## Subplots

```python
# First subplot - Original Price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.title('Original prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Second subplot - Predicted Price
plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Predicted prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Third subplot - Difference between Original & Predicted Prices
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days, price_difference, color='purple', label='Price Difference');
plt.title('Difference between Original & Predicted Prices');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend();

plt.tight_layout();
```

## Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();
plt.savefig('example_prediction.png', dpi=300);
```
